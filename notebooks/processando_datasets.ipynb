### 🔄 Bloco 5 Comparação de Estruturas

```python
colunas_df2022 = set(df2022.columns)
colunas_df2023 = set(df2023.columns)

colunas_df1_nao_df2 = list(colunas_df2022 - colunas_df2023)
colunas_df2_nao_df1 = list(colunas_df2023 - colunas_df2022)

tipos_diferentes = {}
colunas_em_comum = list(colunas_df2022.intersection(colunas_df2023))
for coluna in colunas_em_comum:
    tipo_df2022 = df2022[coluna].dtype
    tipo_df2023 = df2023[coluna].dtype
    if tipo_df2022 != tipo_df2023:
        tipos_diferentes[coluna] = (tipo_df2022, tipo_df2023)


print("Colunas em df2022 mas não em df2023:", colunas_df1_nao_df2)
print("Colunas em df2023 mas não em df2022:", colunas_df2_nao_df1)
print("Colunas com tipos diferentes:")
for coluna, (tipo_df2022, tipo_df2023) in tipos_diferentes.items():
    print(f"  {coluna}: df2022 - {tipo_df2022}, df2023 - {tipo_df2023}")
```

Verifica compatibilidade entre anos:

Colunas idênticas: Nenhuma diferença encontrada
Tipos diferentes: Q005 (float64 em 2022, int64 em 2023)
Importante para garantir concatenação bem-sucedida

### 📝 Bloco 6 Primeira Análise de Valores Nulos

```python
dfArray = [df2020, df2021, df2022, df2023]
ano = 2020
for df in dfArray:
    nulls = df.isnull().sum().reset_index()
    print(nulls)
    nulls.columns = ['coluna', 'n_nulos']
    nulls.to_csv(f"../data/processed/microdados_enem_{str(ano)}/resumo_nulos{str(ano)}.csv", sep=";", index=False)
    ano += 1
```

### 🧹 Bloco 7 Limpeza e Filtragem dos Dados

```python
df2020 = df2020[df2020['TP_ESCOLA'] != 1]
df2020 = df2020[df2020['SG_UF_ESC'] == 'PE']
df2020 = df2020.dropna(subset=['CO_MUNICIPIO_ESC'])
df2020 = df2020.dropna(subset=['CO_PROVA_CN'])
df2020 = df2020.dropna(subset=['CO_PROVA_CH'])
df2020 = df2020.dropna(subset=['CO_PROVA_LC'])
df2020 = df2020.dropna(subset=['CO_PROVA_MT'])
df2020 = df2020.dropna(subset=['NU_NOTA_CN'])
df2020 = df2020.dropna(subset=['Q001'])

df2021 = df2021[df2021['TP_ESCOLA'] != 1]
df2021 = df2021[df2021['SG_UF_ESC'] == 'PE']
df2021 = df2021.dropna(subset=['CO_MUNICIPIO_ESC'])
df2021 = df2021.dropna(subset=['CO_PROVA_CN'])
df2021 = df2021.dropna(subset=['CO_PROVA_CH'])
df2021 = df2021.dropna(subset=['CO_PROVA_LC'])
df2021 = df2021.dropna(subset=['CO_PROVA_MT'])
df2021 = df2021.dropna(subset=['NU_NOTA_CN'])

df2022 = df2022[df2022['TP_ESCOLA'] != 1]
df2022 = df2022[df2022['SG_UF_ESC'] == 'PE']
df2022 = df2022.dropna(subset=['CO_MUNICIPIO_ESC'])
df2022 = df2022.dropna(subset=['CO_PROVA_CN'])
df2022 = df2022.dropna(subset=['CO_PROVA_CH'])
df2022 = df2022.dropna(subset=['CO_PROVA_LC'])
df2022 = df2022.dropna(subset=['CO_PROVA_MT'])
df2022 = df2022.dropna(subset=['NU_NOTA_CN'])

df2023 = df2023[df2023['TP_ESCOLA'] != 1]
df2023 = df2023[df2023['SG_UF_ESC'] == 'PE']
df2023 = df2023.dropna(subset=['CO_MUNICIPIO_ESC'])
df2023 = df2023.dropna(subset=['CO_PROVA_CN'])
df2023 = df2023.dropna(subset=['CO_PROVA_CH'])
df2023 = df2023.dropna(subset=['CO_PROVA_LC'])
df2023 = df2023.dropna(subset=['CO_PROVA_MT'])
df2023 = df2023.dropna(subset=['NU_NOTA_CN'])

dfArray = [df2020, df2021, df2022, df2023]
for df in dfArray:
    print(df.shape)
```

Aplica filtros:

- Remove registros que tem o tipo de escola como "Não respondeu" (TP_ESCOLA != 1)
- Remove registros em que a escola do participante não é de Pernambuco
- Remove registros sem município da escola
- Remove registros sem códigos de prova
- Remove registros sem notas
- 2020 apenas: Remove registros sem questionário socioeconômico respondido

### 📝 Bloco 8 Segunda Análise de Valores Nulos

```python
dfArray = [df2020, df2021, df2022, df2023]
ano = 2020
for df in dfArray:
    nulls = df.isnull().sum().reset_index()
    print(nulls)
    nulls.columns = ['coluna', 'n_nulos']
    nulls.to_csv(f"../data/processed/microdados_enem_{str(ano)}/resumo_nulos{str(ano)}.csv", sep=";", index=False)
    ano += 1
```

### 🗑️ Bloco 9 Remoção de Colunas

```python
colunas_remover = ['CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT', 'TP_ANO_CONCLUIU']

df2020.drop(colunas_remover, axis=1, inplace=True)
df2021.drop(colunas_remover, axis=1, inplace=True)
df2022.drop(colunas_remover, axis=1, inplace=True)
df2023.drop(colunas_remover, axis=1, inplace=True)
```

Remove colunas desnecessárias que possuem valores diferentes entre os datasets dos anos

###  💾 Bloco 10 Concatenação e Exportação Final

```python
dfArray = [df2020, df2021, df2022, df2023]
df_geral = pd.concat(dfArray, ignore_index=True)
ano = 2020
for df in dfArray:
    df.to_csv(f"../data/processed/microdados_enem_{str(ano)}/MICRODADOS_ENEM_{str(ano)}.csv", sep=";", index=False, encoding="latin1")
    ano += 1

df_geral.to_csv("../data/processed/microdados_enem_geral/MICRODADOS_ENEM_GERAL.csv", sep=";", index=False, encoding="latin1")
```

Etapa final do processamento:

- Concatena todos os anos em único DataFrame (df_geral)
- Salva arquivos individuais processados por ano
